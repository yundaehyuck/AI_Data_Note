### model definition
pre-trained TTS 모델을 불러옵니다.

```python
model_name = TTS.list_models()[0]
# Init TTS
tts = TTS(model_name)
```

### speech synthesis
주어진 텍스트를 입력으로 음성 신호를 합성합니다.

```python
input_text = "This is for the SSAFY audio artificial intelligence track."
wav = tts.tts(input_text, speaker=tts.speakers[0], language=tts.languages[0])
```

**TODO-1**: 합성된 wav를 재생하기 위해선 sample rate 정보가 필요합니다.  
다음의 힌트를 통해 sample rate를 추정하고 wav 신호를 재생하시오.

```python
"""
Hint: 아래 코드는 tts.tts 함수의 마지막 실행 코드입니다. 
# process_time = time.time() - start_time
# audio_time = len(wavs) / self.tts_config.audio["sample_rate"]
# print(f" > Processing time: {process_time}")
# print(f" > Real-time factor: {process_time / audio_time}")
"""

process_time = 2.0756564140319824
real_time_factor = 0.44879057600691513

audio_time = process_time/real_time_factor

sample_rate = len(wav)/audio_time

print(sample_rate)
```

```python
ipd.Audio(wav, rate=sample_rate)
```
