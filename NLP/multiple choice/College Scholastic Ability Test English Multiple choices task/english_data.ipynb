# data indexing

```python
data_dict['category'] = []

for q in data_dict['question']:

    
    if '목적' in q:
        
        data_dict['category'].append('목적')
    
    elif '주장' in q:

        data_dict['category'].append('주장')
    
    elif '요지' in q:

        data_dict['category'].append('요지')
    
    elif '주제' in q:

        data_dict['category'].append('주제')
    
    elif '제목' in q:
        
        data_dict['category'].append('제목')
    
    elif '_' in q:
        
        data_dict['category'].append('빈칸')
    
    elif '일치' in q:

        data_dict['category'].append('일치')

    elif '심경' in q:

        data_dict['category'].append('심경')
    
    elif '어법' in q:

        data_dict['category'].append('어법')
    
    elif '문맥' in q:

        data_dict['category'].append('문맥')
    
    elif '전체 흐름' in q:

        data_dict['category'].append('전체 흐름')
    
    else:

        data_dict['category'].append('흐름')

        
```
