# 간단한 사인함수

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

import seaborn as sns ##seaborn
```

```python
#x,y 구성
x = np.linspace(0,14,100)
y1 = np.sin(x)
y2 = 2*np.sin(x+0.5)
y3 = 3*np.sin(x+1.0)
y4 = 4*np.sin(x+1.5)
```

```python
plt.figure(figsize=(10,6))

plt.plot(x,y1, x,y2, x,y3, x,y4)

plt.show()
```

```python
#white grid 적용?

sns.set_style("whitegrid")

plt.figure(figsize=(10,6))
plt.plot(x,y1, x,y2, x,y3, x,y4)
plt.show()
```

# seaborn에서 제공하는 연습용 데이터셋

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

sns.set_style("whitegrid")
%matplotlib inline
```

```python
#요일별로 점심,저녁,흡연여부,식사금액,팁을 정리한 데이터

tips = sns.load_dataset("tips")
tips.head(5)
```

# boxplot

```python
##boxplot
## x축에는 요일, y축에는 전체 금액
#요일별 식사금액

plt.figure(figsize=(8,6))

sns.boxplot(x="day", y="total_bill", data=tips)

plt.show()
```

```python
#hue에서 지정한 범주형 변수 smoker의 여부로,요일 변수를 구분해서 boxplot

plt.figure(figsize=(8,6))

sns.boxplot(x="day", y="total_bill", hue="smoker", data=tips, palette="Set3")

plt.show()
```

# lmplot

```python
#데이터를 scatterplot으로 그리고, 직선으로 regression한 것을 같이 그려줌
#유효범위도 confidence interval로 보여줌

sns.set_style("darkgrid")
sns.lmplot(x="total_bill", y="tip", data=tips, size=7)
plt.show()
```

```python
#hue에 범주형변수를 지정하여, 그에 따라 구분해서 lmplot
#palette는 색상 지정

sns.lmplot(x="total_bill", y="tip", hue="smoker", data=tips, palette="Set1", size=7)

plt.show()
```

# 새로운 연습용 데이터 flights

```python
#연도 및 월별 항공기 승객수를 기록한 데이터

flights = sns.load_dataset("flights")
flights.head(5)
```

```python
#pivot으로 월별, 연도별 구분

flights = flights.pivot("month","year","passengers")
flights.head(5)
```

# heatmap

```python
plt.figure(figsize=(10,8))

sns.heatmap(flights, annot=True, fmt="d")

plt.show()
```

# 새로운 연습용 데이터 iris

```python
sns.set(style="ticks")
iris = sns.load_dataset("iris")
iris.head(10)
```

# pairplot

```python
#꽃잎, 꽃받침의 너비,폭으로 종을 구분

sns.pairplot(iris,hue="species")
plt.show()
```
