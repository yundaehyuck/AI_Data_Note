#1.데이터 정렬

index 혹은 특정 칼럼의 값을 기준으로 행을 오름차순 또는 내림차순으로 정렬할 수 있다

index기준 정렬

df.sort_index(axis = 0, ascending = True, inplace=False)

axis: 0이면 행 인덱스를 기준으로, 1이면 컬럼명을 기준으로 정렬

ascending: True이면 오름차순으로, False이면 내림차순으로 정렬

inplace: True이면 제자리에서 수행 후 None을 반환하고, False이면 복사본을 만들어 정렬을 수행하고 이를 반환

```python
import pandas as pd

from sklearn.datasets import load_iris
```

```python
iris = load_iris()

iris = pd.DataFrame(iris.data, columns = iris.feature_names)
```

```python
iris.head()
```

```python
iris.sort_index(ascending=False, inplace=True) #행 index기준으로 내림차순 정렬함

iris.head()
```

```python
iris.sort_index(axis=1, ascending=True, inplace = True) #컬럼명 기준으로 오름차순 정렬 

iris.head() #알파벳 기준으로 p,q,r,s,.. 오름차순정렬됨
```

ascending: True이면 오름차순으로, False이면 내림차순으로 정렬

inplace: True이면 제자리에서 수행하고 None을 반환하며 False이면 복사본을 만들어 정렬을 수행하고 이를 반환함

#2. 데이터 결합

데이터를 분석하기 위해 두 개 이상의 테이블을 하나의 테이블로 결합해야 하는 경우가 있다

데이터를 결합하는 방법에는 단순 연결(concat, append)과 key를 기준으로 연결하는 조인(join,merge)이 있다

2-1) 단순 연결

단순 연결은 두 개 이상의 데이터프레임을 열 혹은 행 방향으로 그대로 붙이는 것

데이터프레임간 연결 외에도 데이터프레임-Series, Series-Series간에도 연결이 가능

pd.concat(objs, axis=0, ignore_index = False)

objs : concat을 실행할 객체의 리스트(데이터프레임, Series, Panel object)

axis: 0이면 서로를 위 아래로 합치고 1이면 서로를 양옆으로 합친다

ignore_index : True이면 기존 인덱스를 무시하고 0부터 시작하는 정수로 재설정시킴

제목 없음.jpg

먼저 위아래로 결합할 경우에는 각 데이터프레임에서 컬럼의 위치와 관계없이 컬럼명을 기준으로 각 열을 재조립하여 출력함

index는 기본적으로 각 데이터프레임의 고유한 index를 유지하지만 ignore_index=True라고 하면 0부터 1씩 증가하는 정수로 재설정함

```python
import pandas as pd

HR1 = pd.DataFrame({'이름':['장화','홍련'],'부서':['영업','회계'],'직급':['팀장','사원']})

HR2 = pd.DataFrame({'이름':['콩쥐','팥쥐'],'직급':['사원','팀장'],'부서':['영업','인사']})
```

```python
HR1
```

```python
HR2
```

```python
#컬럼 위치가 서로 다르지만 리스트의 첫번째 들어온 데이터프레임의 컬럼 위치를 기준으로 열을 재조립

pd.concat([HR1,HR2],axis=0)
```

```python
pd.concat([HR2,HR1],axis=0)
```

```python
pd.concat([HR1,HR2],axis=0,ignore_index=True)
```

결합하는 두 데이터프레임이 다른 컬럼을 가지고 있는 경우에는 각 데이터프레임의 컬럼을 모두 가져오고 값이 없는 칸은 np.NaN으로 채워진 데이터프레임이 출력

```python
HR3 = pd.DataFrame({'이름':['콩쥐','팥쥐'],'부서':['영업','인사'],'급여':[3500,2800]})

HR3
```

```python
pd.concat([HR1,HR3],axis=0, ignore_index=True)
```

양옆으로 결합할때도 위아래로 결합한것처럼 행 인덱스를 기준으로 데이터프레임을 재조립하여 출력함

결합하는 데이터들의 행 길이가 다른 경우 모든 값을 출력하고 정보가 없는 칸은 np.NaN으로 채운다

```python
HR4 = pd.Series({1:2500},name='급여')

HR4
```

```python
pd.concat([HR1,HR4],axis=1) #series의 2500 index가 1이라서 결합하면 1번 인덱스에 들어감
```

```python
HR5 = pd.DataFrame({'급여':[4500,3000,3500]})

HR5
```

```python
pd.concat([HR1,HR5],axis=1)
```

2. 조인

조인은 연결할 두 데이터프레임이 동시에 가지고 있는 특정한 칼럼 값을 기준으로 2개 이상의 데이터프레임을 결합하여 하나의 데이터프레임으로 만드는 것

두 데이터프레임이 동시에 가지고 있는 컬럼이란, 컬럼명이 같다는 것을 의미하는 것이 아니라 컬럼이 가질 수 있는 값의 집합인 도메인이 같다는 것

이러한 컬럼의 값을 key라고 하며 이를 기준으로 1:1, 1:M, M:M 결합을 할 수 있다

제목 없음.jpg

inner join은 두 데이터프레임 key의 공통된 값을 가져온다

outer join은 두 데이터프레임 key의 값들을 모두 가져오고 각 컬럼에서 없는 값은 NaN으로 표현

left outer join은 왼쪽 데이터프레임의 key 값들을 모두 가져오고 오른쪽 컬럼에서 없는 값은 NaN으로 표현

right outer join은 오른쪽 데이터프레임의 key 값들을 모두 가져오고 왼쪽 컬럼에서 없는 값은 NaN으로 표현

join(), merge() 모두 inner, left, right, outer join을 수행할 수 있지만 merge()가 조금 더 다양한 조건을 사용할 수 있다

df.join(other, on = None, how = 'left', lsuffix-", rsuffix=", sort=False)

df: 결합의 기준이 되는 데이터프레임으로 left dataframe

other: 결합의 대상이 되는 데이터프레임으로 right dataframe

on: key 컬럼으로 두 데이터프레임에서 이름이 같아야함

how: {'left','right','outer','inner'} 중 하나의 조인 방식

lsuffix/rsuffix: 결합한 데이터프레임에서 왼쪽/오른쪽 데이터프레임의 컬럼명에 사용할 접미사

sort: 키 값을 기준으로 정렬할지 여부

df.merge(right, how = 'inner', on = None, left_on = None, right_on = None, left_index=False, right_index=False, sort = False, suffixes = ('_x','_y'))

df: 결합의 기준이 되는 데이터프레임으로 left dataframe

right: 결합의 대상이 되는 데이터프레임으로 right dataframe

how : {'left','right','outer','inner', 'cross'} 조인 수행방식 수행

on: key컬럼의 이름(두 테이블에서 key 컬럼의 이름이 같은 경우에 사용)

left_on/right_on: 두 테이블에서 key 컬럼명이 다를 때 'on'을 대신하여 지정함

left_index/right_index: 왼쪽 혹은 오른쪽 데이터프레임의 인덱스를 key로 사용할 경우 지정

sort: 키 값을 기준으로 정렬할지

suffixes: 결합한 데이터프레임에서 왼쪽/오른쪽 데이터프레임의 컬럼명에 사용할 접미사를 왼쪽, 오른쪽 순서로 각각 지정함

inner join은 교집합 연산과 같다. key 컬럼의 값이 양쪽 데이터프레임에서 모두 존재하는 데이터에 대해서만 조인을 수행한다

어느 한쪽 데이터프레임에만 존재하는 key를 가지고 있는 행은 버려지게 된다

```python
import pandas as pd

product = pd.DataFrame({'상품코드':['G1','G2','G3','G4'],'상품명':['우유','감자','빵','치킨']})

sale = pd.DataFrame({'주문번호':[1001,1002,1002,1003,1004],'상품코드':['G4','G3','G1','G3','G5'],'주문수량':[1,4,2,2,3]})
```

```python
product
```

```python
sale
```

하나의 제품에 대해 여러 건의 주문이 들어올 수 있으므로 1:N 결합이 수행된다

상품과 주문이 모두 존재하는 데이터만 조인하여 출력해보자.

주문되지 않은 상품인 감자와 존재하지 않은 상품 G5에 대한 건은 삭제된다

```python
sale.merge(product, on = '상품코드',how = 'inner')
```

outer join은 full outer join이라고도 하며 합집합 연산과 같다.

조인을 수행하는 양쪽 데이터프레임에 있는 모든 데이터가 조인에 사용되고 출력된다

key컬럼의 값이 양쪽 모두에 존재하는 데이터에 대한 조인은 inner join과 같지만

어느 한쪽에만 존재하는 key의 경우 key가 존재하지 않는 데이터프레임 부분에는 np.NaN 값을 채운다

outer join을 수행하고 key 컬럼인 상품코드를 기준으로 정렬해보자.

1004번 주문의 상품명과 g2 상품의 주문번호와 주문 수량이 NaN으로 표기된다

```python
sale.merge(product, on = '상품코드', how = 'outer', sort=True)
```

right join과 left join은 right outer join, left outer join이라고도 하며, 방향만 다를 뿐 같은 조인 방식이다.

양쪽 데이터프레임에서 하나가 기준이 되는 조인 방식으로 기준이 되는 데이터프레임은 전체가 사용되지만 대상이 되는 데이터프레임은 전체 데이터가 출력되지 않을 수 있다는 것이 특징

대상 데이터프레임에 같은 key값이 없다면 np.NaN을 채운다

모든 주문에 대해 상품명 컬럼을 결합하는 것을 left 조인으로 연습해보자

이번에는 양족 데이터프레임의 컬럼명을 각각 지정해본다.

1004번 주문의 상품명이 NaN으로 표기되었고, 주문이 존재하지 않는 G2 상품에 대한 정보는 삭제된다

```python
#left_on과 right_on 각각에 써도 되지만

#서로 같으면 on 하나에만 써도 된다
sale.merge(product, left_on = '상품코드', right_on = '상품코드', how='left')
```

```python
#기타 옵션 확인용

#suffixes = ('_x','_y') 는 key가 아닌 컬럼인데 컬럼명이 서로 같은 경우가 왼쪽, 오른쪽에 존재하면

#아래와 같이 구분하기 위해 사용됨

sale.merge(product, left_index=True,right_index=True, how = 'inner')
```

```python
#index를 key값으로 사용하고 싶은 경우

sale.merge(product, left_index=True, right_index=True, how='outer')
```

```python
#index를 key로 쓰고싶으면 left_index, right_index에 True를 모두 써줘야함

sale.merge(product,right_index=True, how='right')
```

```python
sale.merge(product,left_index=True, how='left')
```
