# dataframe 병합

```python
df1 = pd.DataFrame({'A':['A0','A1','A2','A3'],
                    'B':['B0','B1','B2','B3'],
                    'C':['C0','C1','C2','C3'],
                    'D':['D0','D1','D2','D3']},
                   index = [0,1,2,3])

df2 = pd.DataFrame({'A':['A4','A5','A6','A7'],
                    'B':['B4','B5','B6','B7'],
                    'C':['C4','C5','C6','C7'],
                    'D':['D4','D5','D6','D7']},
                   index=[4,5,6,7])

df3 = pd.DataFrame({'A':['A8','A9','A10','A11'],
                    'B':['B8','B9','B10','B11'],
                    'C':['C8','C9','C10','C11'],
                    'D':['D8','D9','D10','D11']},
                   index=[8,9,10,11])
```

# concat

열방향(위아래로 합친다)

```python
result = pd.concat([df1,df2,df3])

result
```

```python
#keys 옵션을 주면 df1,df2,df3를 구분하면서 합칠수 있다
#key로 지정된 구분은 다중 index가 되어 level을 형성

result = pd.concat([df1,df2,df3],keys=['x','y','z'])

result
```

```python
result.index
```

```python
#level0의 index 불러오기

result.index.get_level_values(0)
```

```python
#level 1의 index 불러오기

result.index.get_level_values(1)
```

# axis concat

```python
df4 = pd.DataFrame({'B':['B2','B3','B6','B7'],
                    'D':['D2','D3','D6','D7'],
                    'F':['F2','F3','F6','F7']},
                   index=[2,3,6,7])

result = pd.concat([df1,df4],axis=1)
```

# outer join

```python
#axis=1은 행방향으로 병합하는 것

#index를 기준으로 합친다
#df1은 0,1,2,3이고 df2는 2,3,6,7이므로 2,3을 기준으로 서로 합치고
#서로 없는 index에는 NaN으로 채워진다

result
```

# inner join

```python
#공통된 index부분만 합쳐서 가져오고 나머지는 버린다

result = pd.concat([df1,df4],axis=1, join='inner')

result
```

```python
#df1의 index에 맞게 df4의 index 부분은 지워버림

result = pd.concat([df1,df4],axis=1).reindex(df1.index)

result
```

# merge

두 데이터프레임에서 공통된 컬럼인 key를 기준으로 병합하는 명령

on= 옵션에 key 컬럼을 명시

기본적으로 행방향, (양옆으로) 합친다

```python
#inner join

pd.merge(left,right,on='key')
```

how 옵션을 주어서 한쪽 데이터는 전부 가져오고, 나머지 데이터프레임에서 빈 부분은 NaN으로 채우는 join

```python
#left outer join

pd.merge(left,right,how='left',on='key')
```

```python
#right outer join

pd.merge(left,right,how='right',on='key')
```

```python
#full outer join

#두 데이터프레임의 데이터를 모두 가져오는데, 공통된 요소가 아닌 부분은 NaN으로 채워짐

pd.merge(left,right,how='outer',on='key')
```

```python
#outer말고 inner를 주면 처음에 한 inner join

pd.merge(left,right,how='inner',on='key')
```
