#5. index 확인, 추가, 리셋

데이터프레임은 index와 column(속성)을 가지고 있고 이를 통해 인덱스나 컬럼을 수정하거나 출력

인덱스나 컬럼을 수정할 경우 list나 numpy array형식을 사용할 수 있고 기존 데이터와 길이가 같아야함

```python
import pandas as pd

df = pd.DataFrame({'class':['kor','math'],'score':[70,80]})

df
```

df.index로 데이터프레임 df의 index 열을 출력함

```python
df.index
```

```python
list(df.index)
```

df.index = (리스트)로 원하는 인덱스로 변경가능

```python
df.index = ['A','B']
```

```python
df.index
```

```python
df
```

set_index()를 사용하면 데이터프레임내의 열을 인덱스로 변경

set_index(keys, drop=True, append=False, inplace=True)

keys: 인덱스로 사용하고자 하는 컬럼의 이름

drop: 인덱스로 세팅한 컬럼을 데이터프레임 내에서 삭제할지 결정

append: 기존에 존재하던 인덱스를 삭제할지 인덱스로 그대로 둘지

inplace: 원본 객체를 변경할지

```python
df.set_index('class',drop=True, append=False)
```

```python
df.set_index('class',drop=False,append=False)
```

```python
df.set_index('class',drop=True, append=True)
```

```python
df.set_index('class',drop=True, append=True).index
```

inplace를 설정하지않으면 원본 객체를 그대로 둔다

```python
df.set_index('class',drop=True,append=False)
```

```python
df
```

```python
df.set_index('class',drop=True,append=False,inplace=True)
```

```python
df
```

reset_index()는 인덱스를 0부터 시작하는 정수로 재설정함

reset_index(drop=False, inplace=False)

drop: 기존 인덱스를 데이터프레임 내에서 삭제할지, 컬럼으로 만들지

inplace: 원본 객체를 변경할지

```python
df.reset_index(drop=False, inplace=False)
```

```python
df.reset_index(drop=True, inplace=False)
```

필요에 따라 데이터 타입을 변경해야할 필요가 있다

파이썬에서 다음 줄로 코드를 넘기고 싶을 때 \를 추가하면 다음 줄로 코드를 넘길 수 있다

```python
iris[['sepal_width','petal_length']] = 
iris[['sepal_width','petal_length']].astype('int')
```

```python
iris['sepal_length'] = iris['sepal_width'].astype('int')

iris[['sepal_width','petal_length']] = \ #\추가하여 다음 줄로 넘김
iris[['sepal_width','petal_length']].astype('int')
```
